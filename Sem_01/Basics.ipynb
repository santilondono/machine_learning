# Let's start on the basics. Variables and arrays

### First, import some libraries

```
import numpy as np
```

```
array_1 = np.array([5,6,2,6,8,3,5,7,1,4]) # Create an array of N values
array_2 = np.random.randint(1,9,10) # Create an array of random int values
            # with the same lenght as array_1

print(array_1,array_2) # print both array_1 and array_2
print(array_1.shape, array_2.shape)# print both array_1 and array_2 shapes
print(array_1 + array_2) #print summation of array_1 and array_2
```

```
array_1[3] = 70 # Replace the value of array_1 in pos 3 by 70
print(array_1 + array_2) # Print summation again
```

```
array_3 = array_1 * array_2 # Apply element-wise multiplication among array_1 and array_2
print(array_3) # Print results
```

```
for i in range(0,10):
  array_3[i] = array_1[i] * array_2[i]
print(array_3)
```
