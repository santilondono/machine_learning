# Nonlinearities

```
import time
from tqdm import tqdm
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from sklearn.datasets import make_moons, make_blobs


import torch
from torch import nn
from torchsummary import summary
```

```
# Configuration options
num_classes = 2

X, y = make_moons(n_samples= 1000 , noise=0.05)
# X, y = make_blobs(n_features = 2, n_samples= 1000 , centers=num_classes, random_state=1)

X_train, X_test, y_train, y_test = train_test_split( X, y , test_size = 0.20, random_state = 1)
print (X_train.shape, X_test.shape, y_train.shape, y_test.shape)

_, axes = plt.subplots (1,2, figsize=(10,5))

axes[0].scatter( X[:,0] , X[:,1] , c = y , cmap='rainbow', edgecolors='k')
axes[1].scatter( X_train[:,0] , X_train[:,1] , c="m",edgecolors='k')
axes[1].scatter( X_test[:,0] , X_test[:,1] , c="c", edgecolors='k')
```

$\Large{y = mx + b + K(x), K(x) = x \cdot x^T + c}$

```
# # Convert into greyscale
# X_train = X_train.astype('float32')
# X_test = X_test.astype('float32')
```

```
# Convert target classes to categorical ones
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

# Set the input shape
input_shape = (X_train.shape)
print(f'Feature shape: {input_shape}')
```

```
print (X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Create a graph model

```
model = nn.Sequential(
    nn.Linear(2, 10), nn.ReLU(),
    nn.Linear(2, 10), nn.ReLU(),
    nn.Linear(2, 10), nn.ReLU(),
    nn.Linear(10, num_classes)
)

#model.cuda()
summary(model, input_size=(2,),device='cpu')
# print (model)
```

## Define loss function and optimizer

```
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 1e-2)

# model.cuda()
# loss.cuda()
```

## Train model

```
epochs = 1000
batch_size = 100
losses = []

start_time = time.time()
progress = tqdm(range( epochs ), ncols=110)

for epoch in progress:
  #
  batch_losses = 0

  for batch_i in range(0, len(X_train), batch_size):
    #
    batch_X = X_train[ batch_i : batch_i + batch_size ]
    batch_y = y_train[ batch_i : batch_i + batch_size ]
    batch_X = torch.from_numpy(batch_X).to(torch.float32)#.cuda()
    batch_y = torch.from_numpy(batch_y).to(torch.float32)#.cuda()

    # Zero the gradients
    optimizer.zero_grad()

    # Perform forward pass
    predictions = model( batch_X )

    # Compute loss
    batch_loss = loss( batch_y , predictions )

    # Perform backward pass
    batch_loss.backward()

    # Optimize parameters
    optimizer.step()

    ## Save stats
    batch_losses += batch_loss.item()

    elapsed_time = time.time() - start_time

  progress.set_description("Epoch [%d/%d] [Loss: %f] time: %3f" % (epoch, epochs,
                                                                   batch_loss.item(),
                                                                   elapsed_time))

  losses.append(batch_losses/(len(X_train)/batch_size))
```

## Plot results

```
import seaborn as sns

plt.plot(np.array(losses))
plt.xlabel("Epoch")
```

```
h_min, h_max = X_test[:,0] .min()-0.2, X_test[:,0] .max()+0.2
v_min, v_max = X_test[:,1].min()-0.1, X_test[:,1] .max()+0.1

# Crear el grid de las fronteras de decision
h_grid, v_grid = np.meshgrid(np.linspace( h_min , h_max , 1000), np.linspace( v_min , v_max , 1000))

# Predecir los valores del grid
grid_ = torch.from_numpy( np.c_[h_grid.ravel(),v_grid.ravel()] ).to(torch.float32)#.cuda()
pred_grid = model( grid_ )
pred_grid = np.argmax( pred_grid.detach().numpy() , axis=1)
print (pred_grid.shape)

# Encajar las predicciones al tamaÃ±o del grid
pred_grid = pred_grid.reshape(h_grid.shape)
print (pred_grid.shape)
```
